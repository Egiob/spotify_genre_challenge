```python
import bs4
import numpy as np
import urllib
from time import sleep
from multiprocessing.dummy import Pool as ThreadPool


class GenreScrapper:
    def __init__(self):
        self.url = "http://everynoise.com/everynoise1d.cgi?scope=all"
        genres = self.get_genres(self.url)
        genres = list(map(self.parse_genre, genres))
        n = len(genres)
        self.genre_dict = {i: genres[i][0] for i in range(n)}
        self.index_dict = {genres[i][0]: i for i in range(n)}
        self.url_dict = {genres[i][0]:genres[i][1] for i in range(n)}

        self.adj = np.empty((n,n))

    def get_genres(self, url):
        res = urllib.request.urlopen(url).read()
        soup = bs4.BeautifulSoup(res)
        return soup.findAll(attrs={'valign': 'top'})

    def parse_genre(self, genre):
        genre = genre.find(
            attrs={'title': 'Re-sort the list starting from here.'})
        link = genre.get('href')
        return genre.text, link
        
    def get_similarity(self, genre):
        url = "http://everynoise.com/everynoise1d.cgi"+self.url_dict[genre]
        print(url)
        genres = self.get_genres(url)
        genres = list(map(self.parse_genre, genres))
        i = self.index_dict[genre]
        for j in range(len(genres)):
            k = self.index_dict[genres[j][0]]
            self.adj[i, j] = self.f_w(k)
            
    def build_adj_matrix(self):
        pool = ThreadPool(16)
        pool.map(self.get_similarity, self.index_dict.keys())
            
    
    def f_w(self, index):
        return 1/(index+1)
```
